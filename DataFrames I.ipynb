Fill missing values

```python
nba = pd.read_csv("nba.csv").dropna(how='all')
```

```python
nba['Salary'] = nba['Salary'].fillna(0)
```

```python
nba
```

```python
nba['College'] = nba['College'].fillna(value='Unkown')
```

The astype Method 1 (change for example the float to int)

```python
nba = pd.read_csv("nba.csv").dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0)
nba['Weight'] = nba['Weight'].fillna(0)
```

```python
nba.dtypes
```

```python
nba['Salary'] = nba['Salary'].astype(int)
nba['Weight'] = nba['Weight'].astype(int)
nba.dtypes
```

The astype Method 2 (Declare the category to reduce the memory usage, for example there are 30 nba teams so you can specify it)

```python
nba = pd.read_csv("nba.csv").dropna(how ="all")
```

```python
nba["Team"].nunique()
nba.nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
nba["Team"] = nba["Team"].astype("category")
```

```python
nba.dtypes
```

Rank values method

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
nba
```

```python

nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype(int)
```

```python
nba
```

```python
nba.sort_values('Salary Rank', ascending=True).head(10)
```
